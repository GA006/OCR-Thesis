# Training Observation
How to take the log file from the training process and visualize the loss.

```python
PATH_TO_LOG_FILE = './outputs/prima/mask_rcnn_R_50_FPN_3x/log.txt'
```

```python
f = open(PATH_TO_LOG_FILE)
data = f.readlines()
```

```python
iterations = []
losses = []
for i in data:
    if 'total_loss' in i:
        curr = i.split(' ')
        for j in range(len(curr)):
            if curr[j] == 'iter:':
                iterations.append(float(curr[j+1]))
            elif curr[j] == 'total_loss:':
                losses.append(float(curr[j+1]))
            else:
                continue

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6), tight_layout=True)

plt.plot(iterations, losses, 'b-', linewidth=1)

plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend(title='Models', title_fontsize=13, labels=['Mask R-CNN'])
```
